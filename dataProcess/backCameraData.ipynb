```python
# Save depth images
for folder in os.listdir(os.getcwd() + "/data/images"):
    print(folder)
    for filename in os.listdir(os.getcwd() + "/data/images/" + folder):
        print(filename)
        img = cv2.imread('data/images/' + folder + "/" + filename,0)
        dirs = os.getcwd() + "/data/depth/" + folder
        try:
            os.makedirs(dirs)
        except IOError:
            print("dictionary " + dirs + " exist")
        numpy.savetxt(dirs + "/" + filename[0:-5] + ".csv", img, delimiter=",")
```

```python
# Save depth data as csv file
img = cv2.imread('data/images/video0/image-00001.jpeg',0)
name = "video0"
numpy.savetxt("data/depth/" + name[0:-5] + ".csv", img, delimiter=",")
rows,cols = img.shape
print(img)
for i in range(rows):
    print("row" + str(i))
    for j in range(cols):
        k = img[i,j]
        print(k, end=" ")
```
